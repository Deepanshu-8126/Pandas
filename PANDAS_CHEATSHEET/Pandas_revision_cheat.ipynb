# 📅 DAY 1 — Series
### Concept: Series = 1D labelled array. Ek column ka data.
### Analogy: Excel ki sirf ek column — lekin index ke saath.

````python
# ═══════════════════════════════════════
# DAY 1.1 — Series ways 5
# ═══════════════════════════════════════

# 1️⃣ List se
s1 = pd.Series([85, 90, 78, 92, 88])
print('1️⃣ List to  Series:')
print(s1)

# 2️⃣  with  Custom Index
s2 = pd.Series([85, 90, 78, 92, 88],
               index=['Math','Science','English','Hindi','SST'])
print('\n2️⃣ Custom Index  Series:')
print(s2)

# 3️⃣ Dictionary
s3 = pd.Series({'Apple': 150, 'Mango': 80, 'Banana': 40})
print('\n3️⃣ Dictionary to Series:')
print(s3)

# 4️⃣ Scalar value
s4 = pd.Series(5, index=[0, 1, 2, 3])
print('\n4️⃣ Scalar value (same value in all ):')
print(s4)

# 5️⃣ NumPy array
s5 = pd.Series(np.array([10, 20, 30, 40]))
print('\n5️⃣ NumPy array :')
print(s5)
````

````python
# ═══════════════════════════════════════
# DAY 1.2 — Series Indexing & Access
# ═══════════════════════════════════════
# RULE: Label slice → LAST INCLUDE hota hai
#       Integer slice → LAST EXCLUDE hota hai

marks = pd.Series([85, 90, 78, 92, 88],
                  index=['Math','Science','English','Hindi','SST'])

print('Single element (label)    :', marks['Math'])          # 85
print('Single element (position) :', marks[0])               # 85
print('Multiple labels:')
print(marks[['Math','SST']])                                  # double brackets!
print('\nSlice (label) — LAST INCLUDED:')
print(marks['Math':'Hindi'])                                  # Math to Hindi
print('\nSlice (integer) — LAST EXCLUDED:')
print(marks[0:3])                                             # 0,1,2 only
````

````python
# ═══════════════════════════════════════
# DAY 1.3 — Series Attributes
# ═══════════════════════════════════════
# Attributes = properties (no brackets, just dot)

s = pd.Series([85, 90, 78, 92, 88],
              index=['Math','Science','English','Hindi','SST'])
s.name = 'Student_Marks'

print('values :', s.values)    # NumPy array
print('index  :', s.index)     # Index labels
print('dtype  :', s.dtype)     # int64
print('shape  :', s.shape)     # (5,)
print('size   :', s.size)      # 5
print('name   :', s.name)      # Student_Marks
````

````python
# ═══════════════════════════════════════
# DAY 1.4 — Series Operations
# ═══════════════════════════════════════
# KEY RULE: Operations element-wise hote hain — loop ki zarurat nahi!

s = pd.Series([10, 20, 30, 40, 50])

print('Original      :', s.values)
print('s + 5         :', (s + 5).values)
print('s * 2         :', (s * 2).values)
print('s ** 2        :', (s ** 2).values)
print('s > 25        :', (s > 25).values)   # Boolean Series
print('Filter s > 25 :')
print(s[s > 25])                            # Sirf > 25 wale

# Two Series operation
s1 = pd.Series([10, 20, 30], index=['a','b','c'])
s2 = pd.Series([5,  10, 15], index=['a','b','c'])
print('\ns1 + s2:')
print(s1 + s2)

# Mismatched index → NaN
s3 = pd.Series([10, 20], index=['a','b'])
s4 = pd.Series([5,  15], index=['a','c'])  # 'c' alag
print('\nMismatched index → NaN:')
print(s3 + s4)
````

````python
# ═══════════════════════════════════════
# DAY 1.5 — Series Important Methods
# ═══════════════════════════════════════

scores = pd.Series([85, 92, np.nan, 78, 95, 88, np.nan, 70],
                   index=['Ali','Raj','Sam','Tom','Priya','Dev','Mia','Zoe'])

print('--- describe() ---')
print(scores.describe())

print('\n--- NaN count ---')
print('isnull().sum():', scores.isnull().sum())   # 2

print('\n--- fillna() with mean ---')
print(scores.fillna(scores.mean()).round(2))

print('\n--- Top scorer ---')
print('Name:', scores.idxmax(), '| Score:', scores.max())

print('\n--- Filter > 85 ---')
print(scores[scores > 85])

print('\n--- sort_values() descending ---')
print(scores.sort_values(ascending=False))

print('\n--- apply() — Grade assign ---')
def get_grade(x):
    if pd.isna(x):  return 'N/A'
    elif x >= 90:   return 'A'
    elif x >= 80:   return 'B'
    else:           return 'C'
print(scores.apply(get_grade))

# value_counts example
categories = pd.Series(['cat','dog','cat','bird','dog','cat'])
print('\n--- value_counts() ---')
print(categories.value_counts())
````

# 📅 DAY 3 — loc, iloc, Boolean Filtering, query()
### Concept: Data mein se exact cheej nikalna 3 tarikon se.
| Method | Based On | Last Value in Slice |
|--------|----------|---------------------|
| loc | Label | INCLUDE |
| iloc | Position | EXCLUDE |
| Boolean | Condition | — |

````python
# ═══════════════════════════════════════
# DAY 3.1 — loc[] — Label Based
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Single row
print('loc[0] — Row 0:')
print(df.loc[0])

# Row range — LAST INCLUDED
print('\nloc[0:2] — Rows 0,1,2 (last included):')
print(df.loc[0:2, ['Name','Age','Survived']])

# Single cell
print('\nloc[0, Name] — Single cell:')
print(df.loc[0, 'Name'])

# Row + multiple cols
print('\nloc[0, Name to Age]:')
print(df.loc[0, 'Name':'Age'])

# All rows, one col
print('\nloc[:, Age] — All rows, Age column:')
print(df.loc[:, 'Age'].head())
````

````python
# ═══════════════════════════════════════
# DAY 3.2 — iloc[] — Position Based
# ═══════════════════════════════════════

# Single row by position
print('iloc[0] — First row:')
print(df.iloc[0])

# Row range — LAST EXCLUDED
print('\niloc[0:3] — Rows 0,1,2 only (3 excluded):')
print(df.iloc[0:3, [0,2,4]])   # rows 0-2, cols 0,2,4

# Last row
print('\niloc[-1] — Last row:')
print(df.iloc[-1, [2,3,4,5]])  # last row, select cols

# Specific rows + cols
print('\niloc[[0,2,4], [2,3,4]] — Specific rows+cols:')
print(df.iloc[[0,2,4], [2,3,4]])

# Last 3 rows, last 2 cols
print('\niloc[-3:, -2:] — Last 3 rows, last 2 cols:')
print(df.iloc[-3:, -2:])
````

````python
# ═══════════════════════════════════════
# DAY 3.3 — Boolean Filtering
# ═══════════════════════════════════════
# RULE: & = AND,  | = OR,  ~ = NOT
# RULE: Har condition ko () mein wrap karo!

# Single condition
print('1st class passengers:', len(df[df['Pclass'] == 1]))
print('Female passengers   :', len(df[df['Sex'] == 'female']))
print('Age > 50            :', len(df[df['Age'] > 50]))

# AND condition
female_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
print('\nFemale + Survived:', len(female_survived))

# OR condition
class1_or_survived = df[(df['Pclass'] == 1) | (df['Survived'] == 1)]
print('1st class OR Survived:', len(class1_or_survived))

# NOT condition
not_female = df[~(df['Sex'] == 'female')]
print('NOT female (males):', len(not_female))

# isin()
upper = df[df['Pclass'].isin([1, 2])]
print('Pclass 1 or 2:', len(upper))

# between()
young = df[df['Age'].between(20, 30)]
print('Age 20-30:', len(young))

# str.contains()
mr_pass = df[df['Name'].str.contains('Mr.')]
print('Mr. passengers:', len(mr_pass))

# Survival rates
female_rate = df[df['Sex']=='female']['Survived'].mean() * 100
male_rate   = df[df['Sex']=='male']['Survived'].mean()   * 100
print(f'\nFemale survival rate: {female_rate:.1f}%')
print(f'Male   survival rate: {male_rate:.1f}%')
````

````python
# ═══════════════════════════════════════
# DAY 3.4 — query() — SQL Style
# ═══════════════════════════════════════

# query() = zyada readable filtering
print('query — Age > 30:')
print(df.query('Age > 30').shape)

print('query — Female AND Survived:')
print(df.query('Sex == "female" and Survived == 1').shape)

print('query — Age between 20 and 30:')
print(df.query('Age between 20 and 30').shape)

print('query — Pclass in [1,2]:')
print(df.query('Pclass in [1, 2]').shape)

# Variable use karna
min_age = 25
print(f'query — Age > {min_age} (variable):')
print(df.query('Age > @min_age').shape)
````

# 📅 DAY 4 — Data Cleaning
### Concept: Real data kabhi perfect nahi hota. Missing values, duplicates, wrong types — sab fix karo.
### Analogy: Raw survey data clean karna — wahi Data Cleaning hai.

````python
# ═══════════════════════════════════════
# DAY 4.1 — Missing Values
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Check missing values
print('--- Missing Values Count ---')
print(df.isnull().sum())

print('\n--- Missing Values % ---')
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
print(missing_pct[missing_pct > 0])

# fillna — Mean (numeric)
df['Age'].fillna(df['Age'].median(), inplace=True)
print('\n✅ Age filled with median:', df['Age'].isnull().sum(), 'nulls left')

# fillna — Mode (categorical)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print('✅ Embarked filled with mode:', df['Embarked'].isnull().sum(), 'nulls left')

# Drop column with too many nulls
df.drop(columns=['Cabin'], inplace=True)
print('✅ Cabin column dropped (77% null)')

# dropna — sirf specific column pe
df_clean = df.dropna(subset=['Age'])
print(f'\ndropna(subset=Age): {df_clean.shape}')

# notna() — null nahi hai check
print('\nnotna() — non-null Age rows:')
print(df[df['Age'].notna()].shape)
````

````python
# ═══════════════════════════════════════
# DAY 4.2 — Duplicates
# ═══════════════════════════════════════

# Duplicate data banao
data = {
    'OrderID': [101, 102, 103, 101, 104, 102],
    'Product': ['Laptop','Phone','Tablet','Laptop','Watch','Phone'],
    'Price'  : [50000, 15000, 25000, 50000, 8000, 15000]
}
df_dup = pd.DataFrame(data)
print('Original data:')
print(df_dup)

print('\nDuplicate rows found:', df_dup.duplicated().sum())
print('Duplicate rows:')
print(df_dup[df_dup.duplicated()])

df_dup.drop_duplicates(inplace=True)
print('\nAfter removing duplicates:')
print(df_dup)
````

````python
# ═══════════════════════════════════════
# DAY 4.3 — Data Types + astype()
# ═══════════════════════════════════════
# dtype: int64, float64, object, bool, datetime64, category

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'].fillna(df['Age'].median(), inplace=True)

print('Before dtype changes:')
print(df[['Survived','Pclass','Sex','Age']].dtypes)

# int → bool
df['Survived'] = df['Survived'].astype(bool)

# int → category (memory efficient)
df['Pclass'] = df['Pclass'].astype('category')
df['Sex']    = df['Sex'].astype('category')

# float → int
df['Age'] = df['Age'].astype(int)

print('\nAfter dtype changes:')
print(df[['Survived','Pclass','Sex','Age']].dtypes)

# Outlier concept
salaries = pd.Series([30000, 35000, 32000, 40000, 38000, 10000000])
print(f'\nOutlier effect:')
print(f'  Mean   : {salaries.mean():>12,.0f}  ← misleading (outlier!)')
print(f'  Median : {salaries.median():>12,.0f}  ← realistic')
````

````python
# ═══════════════════════════════════════
# DAY 4.4 — String Operations + Feature Engineering
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Extract Title from Name
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.')
print('Title value counts:')
print(df['Title'].value_counts())

# Feature Engineering
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
df['Is_Alone']    = (df['Family_Size'] == 1).astype(int)

print('\nFeature Engineering:')
print(df[['Name','SibSp','Parch','Family_Size','Is_Alone']].head(5))

# Column rename all lowercase
df.columns = df.columns.str.lower()
print('\nColumns lowercase:', df.columns.tolist())
````

# 📅 DAY 5 — apply(), map(), sort_values(), rank()
### Concept: Data transform karna — har element ya row pe custom logic lagana.
| Function | Kahan | Kya Karta Hai |
|----------|-------|---------------|
| map() | Series | Har element transform |
| apply() | Series + DataFrame | Complex logic |
| df.map() | DataFrame | Har cell pe function |

````python
# ═══════════════════════════════════════
# DAY 5.1 — map() on Series
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Dict se map()
df['Sex_Code'] = df['Sex'].map({'male': 0, 'female': 1})
print('Dict map — Sex_Code:')
print(df[['Sex','Sex_Code']].head(5))

# Port names
df['Embarked'].fillna('S', inplace=True)
df['Port'] = df['Embarked'].map({'S':'Southampton','C':'Cherbourg','Q':'Queenstown'})
print('\nPort Names:')
print(df[['Embarked','Port']].value_counts())

# Function se map()
def get_generation(age):
    if age < 18:    return 'Minor'
    elif age < 35:  return 'Young Adult'
    elif age < 55:  return 'Middle Aged'
    else:           return 'Senior'

df['Generation'] = df['Age'].map(get_generation)
print('\nGeneration:')
print(df['Generation'].value_counts())

# Lambda map()
df['Fare_INR'] = df['Fare'].map(lambda x: round(x * 105, 2))
print('\nFare_INR (top 3):')
print(df[['Fare','Fare_INR']].head(3))
````

````python
# ═══════════════════════════════════════
# DAY 5.2 — apply() on Series + DataFrame
# ═══════════════════════════════════════
# axis=0 = column-wise (default)
# axis=1 = row-wise (har row ek saath)

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Series pe apply()
df['Survival_Status'] = df['Survived'].apply(
    lambda x: 'Survived' if x == 1 else 'Died'
)
print('Survival Status:')
print(df['Survival_Status'].value_counts())

# Complex function
def categorize_fare(fare):
    if fare < 10:    return 'Budget'
    elif fare < 30:  return 'Economy'
    elif fare < 100: return 'Business'
    else:            return 'Luxury'

df['Fare_Cat'] = df['Fare'].apply(categorize_fare)
print('\nFare Category:')
print(df['Fare_Cat'].value_counts())

# DataFrame apply — axis=1 (row-wise)
def risk_score(row):
    score = 0
    if row['Pclass'] == 3: score += 3
    elif row['Pclass'] == 2: score += 2
    else: score += 1
    if row['Sex'] == 'male': score += 2
    if row['Age'] > 40: score += 1
    return score

df['Risk_Score'] = df.apply(risk_score, axis=1)
print('\nRisk Score (top 5):')
print(df[['Name','Sex','Pclass','Age','Risk_Score']].head(5))

# DataFrame apply — axis=0 (column-wise)
numeric_df = df[['Age','Fare']]
ranges = numeric_df.apply(lambda col: col.max() - col.min())
print('\nColumn ranges (axis=0):')
print(ranges)
````

````python
# ═══════════════════════════════════════
# DAY 5.3 — sort_values(), nlargest(), rank()
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Single column sort
print('Top 5 oldest passengers:')
print(df.sort_values('Age', ascending=False)[['Name','Age']].head(5))

# Multiple columns sort
print('\nSort by Pclass asc, Age desc:')
multi_sort = df.sort_values(
    by=['Pclass','Age'],
    ascending=[True, False]
)
print(multi_sort[['Name','Pclass','Age']].head(5))

# nlargest / nsmallest — shortcut!
print('\nTop 5 fare payers (nlargest):')
print(df.nlargest(5,'Fare')[['Name','Fare','Pclass']])

print('\n5 youngest passengers (nsmallest):')
print(df.nsmallest(5,'Age')[['Name','Age','Pclass']])

# rank()
df['Fare_Rank'] = df['Fare'].rank(ascending=False, method='dense')
print('\nFare ranks (top 5):')
print(df.sort_values('Fare_Rank')[['Name','Fare','Fare_Rank']].head(5))

# reset_index()
survivors = df[df['Survived']==1].reset_index(drop=True)
print(f'\nSurvivors with reset index: {survivors.index[:5].tolist()}')
````

# 📅 DAY 6 — groupby(), agg(), value_counts(), unique()
### Concept: Data ko groups mein baanto, har group pe calculation karo.
### Split → Apply → Combine
### Analogy: School mein section-wise average marks nikalna — yahi groupby hai.

````python
# ═══════════════════════════════════════
# DAY 6.1 — Basic groupby() + Aggregations
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Basic groupby
print('--- Survival by Pclass ---')
print(df.groupby('Pclass')['Survived'].mean().round(3) * 100)

print('\n--- Avg Age by Sex ---')
print(df.groupby('Sex')['Age'].mean().round(2))

print('\n--- Fare sum by Pclass ---')
print(df.groupby('Pclass')['Fare'].sum().round(2))

# All aggregations
print('\n--- Multiple agg on Fare by Pclass ---')
print(df.groupby('Pclass')['Fare'].agg(['sum','mean','min','max','count']).round(2))
````

````python
# ═══════════════════════════════════════
# DAY 6.2 — agg() Custom + Multiple Columns
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Custom named agg
result = df.groupby('Pclass').agg(
    Total_Passengers = ('PassengerId', 'count'),
    Avg_Age          = ('Age',         'mean'),
    Avg_Fare         = ('Fare',        'mean'),
    Survival_Rate    = ('Survived',    'mean')
)
result['Avg_Age']       = result['Avg_Age'].round(1)
result['Avg_Fare']      = result['Avg_Fare'].round(2)
result['Survival_Rate'] = (result['Survival_Rate'] * 100).round(1)
result.columns = ['Passengers','Avg_Age','Avg_Fare','Survival_%']
print('--- Pclass Summary ---')
print(result)

# Multiple group columns
print('\n--- Sex + Pclass survival ---')
multi = df.groupby(['Sex','Pclass'])['Survived'].agg(['mean','sum','count'])
multi['mean'] = (multi['mean'] * 100).round(1)
multi.columns = ['Survival_%','Survived_Count','Total']
print(multi)

# transform() — original shape maintain
df['Pclass_Avg_Fare'] = df.groupby('Pclass')['Fare'].transform('mean')
df['Fare_vs_Class_Avg'] = (df['Fare'] - df['Pclass_Avg_Fare']).round(2)
print('\n--- Fare vs Class Average (transform) ---')
print(df[['Pclass','Fare','Pclass_Avg_Fare','Fare_vs_Class_Avg']].head(8))
````

````python
# ═══════════════════════════════════════
# DAY 6.3 — value_counts(), unique(), nunique()
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# value_counts()
print('--- Pclass distribution ---')
print(df['Pclass'].value_counts())
print('\n--- Pclass % ---')
print((df['Pclass'].value_counts(normalize=True) * 100).round(1))

# unique() and nunique()
print('\n--- unique() ---')
print('Unique Embarked:', df['Embarked'].unique())

print('\n--- nunique() per column ---')
print(df.nunique())

# Per-group unique count
print('\n--- Unique Pclass per Sex ---')
print(df.groupby('Sex')['Pclass'].nunique())
````

# 📅 DAY 7 — merge(), join(), concat()
### Concept: Alag alag DataFrames ko ek mein jodhna.
| Function | Kaise Join | Use Karo Jab |
|----------|------------|---------------|
| concat() | Chipkao (no key) | Same structure tables stack karo |
| merge() | Common column se | SQL-style JOIN |
| join() | Index se | Fast index-based join |

````python
# ═══════════════════════════════════════
# DAY 7.1 — concat()
# ═══════════════════════════════════════
# axis=0 = neeche chipkao (rows badhti hain)
# axis=1 = side mein chipkao (columns badhte hain)

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

movies   = df[df['Pclass'] == 1].copy()  # 1st class
economy  = df[df['Pclass'] == 3].copy()  # 3rd class

print(f'1st class: {movies.shape}')
print(f'3rd class: {economy.shape}')

# axis=0 — neeche stack karo
combined = pd.concat([movies, economy], ignore_index=True)
print(f'Combined:  {combined.shape}')
print('Index:', combined.index[:5].tolist(), '...')

# keys parameter — source track karo
keyed = pd.concat([movies, economy], keys=['1st','3rd'])
print('\nWith keys:')
print(keyed.loc['1st'].head(2))

# axis=1 — side mein chipkao
basic  = df[['PassengerId','Name','Age']]
extras = df[['Survived','Pclass','Fare']]
side   = pd.concat([basic, extras], axis=1)
print(f'\nSide concat (axis=1): {side.shape}')
print(side.head(3))
````

````python
# ═══════════════════════════════════════
# DAY 7.2 — merge() — All 4 Join Types
# ═══════════════════════════════════════
# DRY RUN with simple example

employees = pd.DataFrame({
    'emp_id': [101, 102, 103, 104],
    'name'  : ['Raj', 'Priya', 'Sam', 'Ali'],
    'dept_id': [10, 20, 10, 30]
})

departments = pd.DataFrame({
    'dept_id'  : [10,   20,   40],
    'dept_name': ['IT', 'HR', 'Finance']
})

print('employees:')
print(employees)
print('\ndepartments:')
print(departments)

# INNER — sirf matching rows
inner = pd.merge(employees, departments, on='dept_id', how='inner')
print('\n🔵 INNER JOIN (only matches):')
print(inner)
print('Ali (dept 30) → OUT. Finance (dept 40) → OUT.')

# LEFT — left ki saari rows
left = pd.merge(employees, departments, on='dept_id', how='left')
print('\n🟢 LEFT JOIN (all employees):')
print(left)
print('Ali (dept 30) → STAYS with NaN. Finance → OUT.')

# RIGHT — right ki saari rows
right = pd.merge(employees, departments, on='dept_id', how='right')
print('\n🟡 RIGHT JOIN (all departments):')
print(right)
print('Finance (dept 40) → STAYS with NaN. Ali → OUT.')

# OUTER — dono ki saari rows
outer = pd.merge(employees, departments, on='dept_id', how='outer')
print('\n🔴 OUTER JOIN (everyone):')
print(outer)
print('Ali + Finance → BOTH stay with NaN where no match.')
````

````python
# ═══════════════════════════════════════
# DAY 7.3 — merge() Advanced + indicator
# ═══════════════════════════════════════

employees = pd.DataFrame({
    'emp_id': [101, 102, 103, 104],
    'name'  : ['Raj', 'Priya', 'Sam', 'Ali'],
    'dept_id': [10, 20, 10, 30]
})
departments = pd.DataFrame({
    'dept_id'  : [10, 20, 40],
    'dept_name': ['IT', 'HR', 'Finance']
})

# indicator=True — source track karo
result = pd.merge(employees, departments,
                  on='dept_id', how='outer', indicator=True)
print('indicator=True result:')
print(result)
print('\n_merge column counts:')
print(result['_merge'].value_counts())

# Different column names — left_on / right_on
emp2 = employees.copy()
emp2.rename(columns={'dept_id': 'department_id'}, inplace=True)

result2 = pd.merge(emp2, departments,
                   left_on='department_id',
                   right_on='dept_id',
                   how='inner')
print('\nleft_on/right_on (different column names):')
print(result2)
````

# 📅 DAY 8 — pivot_table(), pivot(), melt(), stack(), unstack()
### Concept: Data ka shape change karna.
```
Long → Wide  :  pivot() / pivot_table()
Wide → Long  :  melt()
Col → Row    :  stack()
Row → Col    :  unstack()
```

````python
# ═══════════════════════════════════════
# DAY 8.1 — pivot() — Basic Reshape
# ═══════════════════════════════════════
# RULE: pivot() mein duplicates nahi hone chahiye → Error aata hai
# RULE: pivot_table() duplicates handle kar leta hai (aggregation se)

# Simple example
sample = pd.DataFrame({
    'Region'  : ['West','West','East','East','South','South'],
    'Category': ['Furniture','Technology','Furniture','Technology',
                 'Furniture','Technology'],
    'Sales'   : [1000, 2000, 1500, 2500, 800, 1200]
})

print('Original (Long Format):')
print(sample)

# pivot — Long → Wide
pivoted = sample.pivot(index='Region',
                       columns='Category',
                       values='Sales')
pivoted.columns.name = None
pivoted.index.name   = None
print('\nAfter pivot() — Wide Format:')
print(pivoted)
````

````python
# ═══════════════════════════════════════
# DAY 8.2 — pivot_table() — Pivot + Aggregation
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Basic pivot_table
print('--- Pclass vs Sex — Survival Mean ---')
pt1 = df.pivot_table(values='Survived',
                     index='Pclass',
                     columns='Sex',
                     aggfunc='mean').round(3)
print(pt1)

# margins=True — Grand Total
print('\n--- With Grand Total (margins=True) ---')
pt2 = df.pivot_table(values='Fare',
                     index='Pclass',
                     columns='Sex',
                     aggfunc='sum',
                     margins=True,
                     margins_name='TOTAL').round(2)
print(pt2)

# Multiple aggfunc
print('\n--- Multiple aggfunc ---')
pt3 = df.pivot_table(values='Fare',
                     index='Pclass',
                     columns='Sex',
                     aggfunc=['sum','mean']).round(2)
print(pt3)

# fill_value=0
print('\n--- fill_value=0 (NaN → 0) ---')
pt4 = df.pivot_table(values='Survived',
                     index='Pclass',
                     columns='Embarked',
                     aggfunc='count',
                     fill_value=0)
print(pt4)
````

````python
# ═══════════════════════════════════════
# DAY 8.3 — melt() — Wide to Long
# ═══════════════════════════════════════
# melt() = pivot() ka ulta
# Columns ko rows mein convert karo

# Wide format
wide = pd.DataFrame({
    'Student' : ['Ali', 'Raj', 'Priya'],
    'Math'    : [85, 92, 78],
    'Science' : [90, 88, 95],
    'English' : [78, 85, 92]
})
print('Wide Format (before melt):')
print(wide)

# melt — Wide → Long
long = wide.melt(
    id_vars   = 'Student',
    value_vars = ['Math','Science','English'],
    var_name   = 'Subject',
    value_name = 'Marks'
)
print('\nLong Format (after melt):')
print(long.sort_values('Student').reset_index(drop=True))

# Monthly sales example
monthly = pd.DataFrame({
    'Product'   : ['Laptop', 'Phone', 'Tablet'],
    'Jan_Sales'  : [50000, 30000, 20000],
    'Feb_Sales'  : [55000, 32000, 22000],
    'Mar_Sales'  : [60000, 35000, 25000]
})

long_monthly = monthly.melt(
    id_vars   = 'Product',
    value_vars = ['Jan_Sales','Feb_Sales','Mar_Sales'],
    var_name   = 'Month',
    value_name = 'Sales'
)
long_monthly['Month'] = long_monthly['Month'].str.replace('_Sales','')
print('\nMonthly Sales — Long Format:')
print(long_monthly.sort_values(['Product','Month']).reset_index(drop=True))
````

````python
# ═══════════════════════════════════════
# DAY 8.4 — stack() and unstack()
# ═══════════════════════════════════════
# stack()   = columns → rows (innermost level)
# unstack() = rows → columns (innermost level)

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# pivot_table banao pehle
pt = df.pivot_table(values='Fare',
                    index='Pclass',
                    columns='Sex',
                    aggfunc='mean').round(2)
print('Pivot Table (Wide):')
print(pt)

# stack() — columns → rows
stacked = pt.stack().reset_index()
stacked.columns = ['Pclass','Sex','Avg_Fare']
print('\nAfter stack() — Long format:')
print(stacked)

# unstack() — rows → columns
multi_idx = df.groupby(['Pclass','Sex'])['Survived'].mean().round(3)
print('\nGroupby result (MultiIndex Series):')
print(multi_idx)
print('\nAfter unstack() — Wide Table:')
print(multi_idx.unstack())
````

# 📅 DAY 9 — DateTime Handling
### Concept: Dates ke saath kaam karna — extract, filter, calculate, resample.
### Rule: CSV se dates STRING mein aati hain. Pehle pd.to_datetime() se convert karo!
```
String '2021-01-15'  →  pd.to_datetime()  →  datetime64
datetime64           →  .dt.year/.month   →  Extract components
date2 - date1        →  Timedelta         →  Days difference
df.resample('M')     →  Monthly grouping  →  Time series analysis
```

````python
# ═══════════════════════════════════════
# DAY 9.1 — pd.to_datetime() + format
# ═══════════════════════════════════════

# Basic conversion
dates = pd.Series(['2021-01-15', '2021-06-20', '2022-03-10'])
print('Before:', dates.dtype)   # object (string)
dates_converted = pd.to_datetime(dates)
print('After :', dates_converted.dtype)   # datetime64
print(dates_converted)

# Different formats
print('\n--- Different formats ---')
print(pd.to_datetime('15/01/2024', format='%d/%m/%Y'))
print(pd.to_datetime('January 15, 2024', format='%B %d, %Y'))
print(pd.to_datetime('15-Jan-2024', format='%d-%b-%Y'))

# errors='coerce' — galat dates → NaT
messy = pd.Series(['2021-01-15', 'not a date', '2021-03-20', None])
print('\nerrors=coerce — galat dates → NaT:')
print(pd.to_datetime(messy, errors='coerce'))

# Format codes reference
print('\n--- Format Codes Reference ---')
ref = {
    '%Y': '4-digit year (2024)',
    '%y': '2-digit year (24)',
    '%m': 'Month number (01-12)',
    '%d': 'Day number (01-31)',
    '%B': 'Month full name (January)',
    '%b': 'Month short name (Jan)',
    '%H': 'Hour 24h (00-23)',
    '%M': 'Minute (00-59)'
}
for code, meaning in ref.items():
    print(f'  {code}  →  {meaning}')
````

````python
# ═══════════════════════════════════════
# DAY 9.2 — .dt Accessor — Components Extract
# ═══════════════════════════════════════
# .dt = datetime accessor — date ke har compartment ki chaabi

# Sample dates
dates = pd.Series(pd.date_range('2022-01-01', periods=6, freq='45D'))
print('Dates:')
print(dates.values)

df_dt = pd.DataFrame({'Date': dates})
df_dt['Year']         = df_dt['Date'].dt.year
df_dt['Month']        = df_dt['Date'].dt.month
df_dt['Month_Name']   = df_dt['Date'].dt.month_name()
df_dt['Day']          = df_dt['Date'].dt.day
df_dt['Weekday_Num']  = df_dt['Date'].dt.dayofweek   # 0=Mon, 6=Sun
df_dt['Weekday_Name'] = df_dt['Date'].dt.day_name()
df_dt['Quarter']      = df_dt['Date'].dt.quarter
df_dt['Day_of_Year']  = df_dt['Date'].dt.dayofyear
df_dt['Is_Month_End'] = df_dt['Date'].dt.is_month_end

print('\n.dt accessor — all components:')
print(df_dt.to_string())
````

````python
# ═══════════════════════════════════════
# DAY 9.3 — Timedelta — Date Difference
# ═══════════════════════════════════════
# Timedelta = do dates ke beech ka time difference
# NaT = Not a Time (DateTime ka NaN)

orders = pd.DataFrame({
    'Order_Date'   : ['2024-01-05', '2024-01-10', '2024-01-15',
                      '2024-01-20', '2024-01-25'],
    'Delivery_Date': ['2024-01-10', '2024-01-12', '2024-01-22',
                      '2024-01-24', '2024-02-01']
})

orders['Order_Date']    = pd.to_datetime(orders['Order_Date'])
orders['Delivery_Date'] = pd.to_datetime(orders['Delivery_Date'])

# Difference
orders['Delivery_Days'] = (orders['Delivery_Date'] -
                           orders['Order_Date']).dt.days

# Expected delivery — 5 din ke andar
orders['Expected']  = orders['Order_Date'] + pd.Timedelta(days=5)
orders['Is_Late']   = orders['Delivery_Date'] > orders['Expected']

print('Delivery Analysis:')
print(orders.to_string())
print(f'\nAvg Delivery Days : {orders["Delivery_Days"].mean():.1f}')
print(f'Late Orders       : {orders["Is_Late"].sum()}')

# pd.Timedelta options
print('\n--- pd.Timedelta Examples ---')
base = pd.Timestamp('2024-01-15')
print(f'+ 7 days   : {base + pd.Timedelta(days=7)}')
print(f'+ 2 weeks  : {base + pd.Timedelta(weeks=2)}')
print(f'+ 12 hours : {base + pd.Timedelta(hours=12)}')
````

````python
# ═══════════════════════════════════════
# DAY 9.4 — Date Filtering
# ═══════════════════════════════════════

df_time = pd.DataFrame({
    'Date' : pd.date_range('2022-01-01', '2023-12-31', freq='W'),
    'Sales': np.random.randint(1000, 10000,
                size=len(pd.date_range('2022-01-01','2023-12-31',freq='W')))
})

# Year filter
y2022 = df_time[df_time['Date'].dt.year == 2022]
print(f'2022 records: {len(y2022)}')

# Quarter filter
q4 = df_time[df_time['Date'].dt.quarter == 4]
print(f'Q4 records: {len(q4)}')

# Date range filter
mask = (df_time['Date'] >= '2022-06-01') & \
       (df_time['Date'] <= '2022-12-31')
h2_2022 = df_time[mask]
print(f'H2 2022 records: {len(h2_2022)}')

# between()
q3 = df_time[df_time['Date'].between('2022-07-01', '2022-09-30')]
print(f'Q3 2022 (between): {len(q3)}')

# Weekday filter
df_time['Weekday'] = df_time['Date'].dt.day_name()
print('\nRecords by weekday:')
print(df_time['Weekday'].value_counts())

# pct_change — growth
yearly_sales = df_time.groupby(df_time['Date'].dt.year)['Sales'].sum()
print('\nYearly Sales:')
print(yearly_sales)
print('YoY Growth %:')
print((yearly_sales.pct_change() * 100).round(2))
````

````python
# ═══════════════════════════════════════
# DAY 9.5 — resample() + date_range() + strftime()
# ═══════════════════════════════════════
# resample() ke liye datetime INDEX zaroori hai!
# Frequency codes: D=Daily, W=Weekly, M=Monthly,
#                  Q=Quarterly, Y=Yearly, B=Business days

# Daily data banao
daily = pd.DataFrame({
    'Sales' : np.random.randint(500, 5000,
                size=len(pd.date_range('2023-01-01','2023-12-31',freq='D'))),
    'Orders': np.random.randint(10, 100,
                size=len(pd.date_range('2023-01-01','2023-12-31',freq='D')))
}, index=pd.date_range('2023-01-01', '2023-12-31', freq='D'))

print(f'Daily data shape: {daily.shape}')

# Monthly resample
monthly = daily.resample('M').sum()
print('\nMonthly Sales (resample M):')
print(monthly.head(6))

# Quarterly
quarterly = daily.resample('Q').agg({'Sales':'sum', 'Orders':'count'})
print('\nQuarterly (resample Q):')
print(quarterly)

# date_range()
print('\n--- date_range() examples ---')
print('Next 6 months (Month Start):')
print(pd.date_range(start='2024-01-01', periods=6, freq='MS'))

print('\nBusiness days in Jan 2024:')
biz = pd.date_range('2024-01-01', '2024-01-31', freq='B')
print(f'Count: {len(biz)}')

# strftime — DateTime → String
print('\n--- strftime() — DateTime to String ---')
ts = pd.Timestamp('2024-03-15')
print('DD/MM/YYYY   :', ts.strftime('%d/%m/%Y'))
print('DD-Mon-YYYY  :', ts.strftime('%d-%b-%Y'))
print('Month Year   :', ts.strftime('%B %Y'))
print('Full Date    :', ts.strftime('%d %B, %Y'))
````

````python
# ═══════════════════════════════════════
# ANSWERS — Pehle khud try karo, phir dekho!
# ═══════════════════════════════════════

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df  = pd.read_csv(url)

# A1
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print('A1 ✅ Missing values filled:', df[['Age','Embarked']].isnull().sum().tolist())

# A2
print('\nA2 ✅ loc[5:10]:')
print(df.loc[5:10, ['Name','Age','Survived']])

# A3
q3 = df[(df['Sex']=='female') & (df['Pclass']==1)]
print(f'\nA3 ✅ Female + 1st class: {len(q3)} passengers')

# A4
def age_group(x):
    if x < 18:   return 'Child'
    elif x < 60: return 'Adult'
    else:        return 'Senior'
df['Age_Group'] = df['Age'].apply(age_group)
print('\nA4 ✅ Age_Group:')
print(df['Age_Group'].value_counts())

# A5
print('\nA5 ✅ Survival rate by Pclass:')
print((df.groupby('Pclass')['Survived'].mean() * 100).round(1))

# A6
print('\nA6 ✅ Pivot Table:')
print(df.pivot_table(values='Survived', index='Pclass',
                     columns='Sex', aggfunc='mean').round(3))

# A7
df['Sex_Code'] = df['Sex'].map({'male':0, 'female':1})
print('\nA7 ✅ Sex encoded:', df['Sex_Code'].value_counts().to_dict())

# A8
print('\nA8 ✅ Top 5 fare payers:')
print(df.nlargest(5,'Fare')[['Name','Fare','Pclass']])

# A9
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
print('\nA9 ✅ Survival by Family Size:')
print((df.groupby('Family_Size')['Survived'].mean() * 100).round(1))

# A10
df['Fare_Rank'] = df['Fare'].rank(ascending=False, method='dense')
print('\nA10 ✅ Top 3 by Fare Rank:')
print(df.nsmallest(3,'Fare_Rank')[['Name','Fare','Fare_Rank']])
````
