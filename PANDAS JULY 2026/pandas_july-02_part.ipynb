## TOPIC - TRANSFORM 

```python
# difference agg() and transform 
# agg - rows decrease 
df.groupby("Category")["Sales"].sum()

# 3 rows
```

```python
# transform rows same category at every label
df['Category_Total'] = df.groupby("Category")['Sales'].transform("sum")
df.sample(3)
```

```python
# sales cateogry distribution
df["Category_Total"] = (
    df.groupby("Category")["Sales"]
      .transform("sum")
)
df["Contribution_%"] = (
    df["Sales"] /
    df["Category_Total"]
) * 100

```

```python
# question - category wise average sales  every row 
df['Avg_Sales']=(df.groupby("Category")["Sales"].transform('mean'))

# question region profit every row 
df['Region_Profit']=(df.groupby("Region")['Profit'].transform("sum"))
```

```python
# rank find 
df['Rank']=df.groupby("Category")['Sales'].rank(ascending=False)
```

```python
## Question - Region wise Sales 
# in pivot table
# syntax - pd.pivot_table(df,values='sales',index = 'region',aggfunc='sum)

pd.pivot_table(df,values='Sales',index="Region",aggfunc="sum")
```

```python
# ques - category wise profit 
pd.pivot_table(df,values="Profit",index="Category",aggfunc="sum")
```

```python
# two dimensions 
# question - region vs category
pd.pivot_table(df,values="Profit",index='Region',columns='Category',aggfunc='sum')
```

```python
# multiple aggregation sales sum + mean
pd.pivot_table(df,values="Sales",index="Category",aggfunc=['sum','mean'])
```

```python
# Fill missing 
pd.pivot_table(df,values="Sales",index='Region',columns='Category',aggfunc='sum',fill_value=0)
```

```python
# margin grand total
pd.pivot_table(
    df,
    values="Sales",
    index="Region",
    columns="Category",
    aggfunc="sum",
    margins=True
)
```

```python
# question - region vs segment unique customers
pd.pivot_table(df,values='Customer Name',aggfunc='nunique',columns='Segment',index='Region')
```

## String function .str

```python
df['Product Name'].str.lower()
df['City'].str.upper()
df['City'].str.title()
df['City'].str.strip().head(4)  
# extra spaces remove

df[df['Product Name'].str.contains("Chair")]
# contains chair product 

df[df['Product Name'].str.contains("chair",case=False)]

# str.len -- lenght count
df['Product Name'].str.len()

# str.replace()
df['City'].str.replace("New York","NY")

# str.split()
# df['director'].str.split(",")

# question 3 - Product names > 40 chars 
df[df['Product Name'].str.len()>40]

# customer name lowercase 
df['Customer Name'].str.lower()

# products containing desk and sales > 500
df[
    df["Product Name"].str.contains(
        "Desk",
        case=False
    )
    &
    (df["Sales"]>500)
]
df.head(3)
```
