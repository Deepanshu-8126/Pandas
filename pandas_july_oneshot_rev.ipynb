```python
# filtering module 
# basic filtering 
# ques - sales greater than 1000
df[df['Sales']>1000]

# loss
df[df['Profit']<0]


# disocunt 
df[df['Discount']>0.2]
```

```python
# text filtering 
df[df['Category']== 'Technology']

# region = west 
df[df['Region']=='West']

# category == furniture 
df[df['Category']=='Furniture'].head(3)
```

```python
# multiple conditions and 
df[(df['Category']=='Technology') & (df['Region']=='West')]


# sales > 1000 and profit > 100
df[(df['Sales']>1000) & (df['Profit']>100)].head(3)

# or condition same | sign
df[(df['Category']=='Technology') | (df['Category']=='Furniture')]

# not condition 
# technology remove 
df[~(df['Category']=='Technology')]

# loss remove 
df[~(df['Profit']<0)]

# east or west and sales > 1000
df[(df['Region']=='West')|(df['Region']=='East') & (df['Sales']>1000)]
```

```python

# module - isin() multiple values check same output  code clean 
df[df['Region'].isin(['West','East'])].tail(10)

df[df['Category'].isin(['Furniture','Technology'])]

# west and east region and sales >1000
#df[df['Region'].isin(['East','West']) &  df(['Sales']>1000)]

# technology / furniture category + profit > 200 + discounnt <0.2
df[(df['Category'].isin(['Technology','Furniture'])) & (df['Profit']>200) & (df['Discount']<0.2)]
```

```python
# between 
# without between 
df[(df['Sales']>=500) & (df['Sales']<=2000)]

# with between  range check 
df[df['Sales'].between(500,1200)]

# sales 100-500
df[df['Sales'].between(100,500)]

# profit 50,300 
df[df['Profit'].between(50,300)]

# sales 1000 - 5000 and profit > 500 
df[ (df['Sales'].between(1000,5000)) & (df['Profit']>500)]


# technology products with salary range 
df[(df['Category']=='Technology') & (df['Sales'].between(500,3000))]
```

```python
# QUERY df.query ( when conditions messy )
# clean syntax 
df.query("Sales > 1000 and Profit > 100 ")


# hard
df.query("Region == 'West' and Sales > 1000 and Profit > 100")

df.query("Region.isin(['East','West']) and Category== 'Technology' ")
```
