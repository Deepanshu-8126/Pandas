```python
# merge
shows = df[['show_id', 'type', 'title', 'release_year']].copy()
# Table 2: Show details
details = df[['show_id', 'rating', 'duration',
              'country', 'listed_in']].copy()


# Table 3: Director info (manually banate hain realistic ke liye)
directors = df[['show_id', 'director']].dropna().copy()

print(f"shows:     {shows.shape}")      # (8807, 4)
print(f"details:   {details.shape}")    # (8807, 5)
print(f"directors: {directors.shape}")  # (6173, 2)

```

```python
# Type 1 - INNer Join
result = pd.merge(shows , details , on = 'show_id', how = 'inner')

print(result.shape)
print(result.head(3))
```

```python
# Type 2 - Left join
# shows = 8807 rows ( left table )
# direcctors = 6173rows missing many
# left join - shows all rows 8807 comes
# in director table - nan comes matching from left table show_idd match their right table comes

result_left = pd.merge(shows , directors , on = 'show_id', how = 'left')

print(result_left.shape)  # shows one table all rowss
print(result_left.isnull().sum())  # director 2634 no direcrtor inffo 2634
```

```python
# Right join
# right table all rows
# from left only take matching

result_right = pd.merge(shows , directors , on = 'show_id' , how = 'right')

# dry run
# directors = 6173 rows right table
# right join : directors all rows will be here
# note show_info not in place nan

print(result_right.shape)
```

```python
# outer join
# both tables all rows
# where match n0t in place of nan comes
result_outer = pd.merge(shows, directors,
                        on='show_id',
                        how='outer')


print(result_outer.shape)
print(result_outer.isnull().sum())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('netflix_titles.csv')

# ── Step 1: Tables banao ──
shows     = df[['show_id','type','title','release_year']].copy()
details   = df[['show_id','rating','duration','country']].copy()
genres    = df[['show_id','listed_in']].copy()
directors = df[['show_id','director']].dropna().copy()

print("Tables created:")
print(f"  shows:     {shows.shape}")
print(f"  details:   {details.shape}")
print(f"  genres:    {genres.shape}")
print(f"  directors: {directors.shape}")

# ── Step 2: Full info merge ──
full = pd.merge(shows, details, on='show_id', how='left')
full = pd.merge(full, genres,   on='show_id', how='left')
print(f"\nFull merged: {full.shape}")

# ── Step 3: Director analysis — only those who have info ──
with_director = pd.merge(shows, directors,
                         on='show_id', how='inner')
print(f"\nShows with director: {with_director.shape}")

# ── Step 4: Top directors ──
top_directors = (with_director.groupby('director')['show_id']
                 .count()
                 .sort_values(ascending=False)
                 .head(10))
print("\nTop 10 Most Prolific Directors:")
print(top_directors)

# ── Step 5: Missing directors ──
missing_dir = pd.merge(shows, directors,
                       on='show_id',
                       how='left',
                       indicator=True)
no_dir = missing_dir[missing_dir['_merge'] == 'left_only']
print(f"\nShows with NO director info: {len(no_dir)}")

# ── Step 6: Movies vs TV Shows breakdown ──
movies_df   = shows[shows['type'] == 'Movie']
tvshows_df  = shows[shows['type'] == 'TV Show']

movies_full = pd.merge(movies_df,  details, on='show_id')
tv_full     = pd.merge(tvshows_df, details, on='show_id')

print("\n--- Movies by Rating ---")
print(movies_full['rating'].value_counts().head(8))

print("\n--- TV Shows by Rating ---")
print(tv_full['rating'].value_counts().head(8))

# ── Step 7: Country analysis ──
country_count = (full.groupby('country')['show_id']
                 .count()
                 .sort_values(ascending=False)
                 .head(10))
print("\nTop 10 Countries by Content:")
print(country_count)

# ── Step 8: Year-wise content addition ──
yearly = (full.groupby(['release_year','type'])['show_id']
          .count()
          .unstack()
          .fillna(0)
          .astype(int))
print("\nContent added per year (last 10 years):")
print(yearly.tail(10))
```
