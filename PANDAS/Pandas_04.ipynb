```python
# setup
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Titanic-Dataset.csv')
print(df.shape)  # 891 rows and 12 columns
print(df.isnull().sum()) # total count of null values



# age - 177 missing , cabin = 687 missing too much missing
```

```python
# dropna : rows column drop
df_clean = df.dropna()
print(df_clean.shape)  # 183 12 many rows remove


# specific column mey null so drop
df_clean = df.dropna(subset = ['Age'])
df_clean.shape  # 714 12

# if 2 and more than  null in rows then drop
df_clean =  df.dropna(thresh=10)  #miniumn 10 non values need
df_clean.shape  # 891

# column drop if very null
df_clean = df.dropna(axis=1)
df_clean.shape # 891 9


# rules : if 40 percent more than null valeus in ccolumn then column drop beeter
```

```python
# fillna  : fill the values

# specific value to fill 0 to fill
df['Age'].fillna(0,inplace=True) # bad practice

# mean to fill common
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Age'].head(10)


# median to fill
df['Age'].fillna(df['Age'].median(),inplace=True)
# median - values sort middle one take


# mode to fill  - for categorial data
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)


# Forward fill - first valeu comes from the null value
df['Age'].fillna(method='ffill', inplace=True)


# backward fill
df['Age'].fillna(method='bfill', inplace=True)
```

```python
# 3 way- interpolation
# estimate middle value mathematically time series data use

df['Age'].interpolate(method='linear',inplace = True)

# domian knowledge fill
df['Cabin'].fillna('Unknown', inplace=True)
```

```python
# complete data cleaniing pipeline
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Titanic-Dataset.csv')
print(f"Original shape: {df.shape}")


# ── Step 2: Initial Inspection ──
print(df.info())
print(df.isnull().sum())
print(df.duplicated().sum())


# Step 3: Drop useless columns ──
df.drop(columns=['PassengerId', 'Ticket', 'Cabin'], inplace=True)


# ── Step 4: Handle Missing Values ──
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)


# ── Step 5: Remove Duplicates ──
df.drop_duplicates(inplace=True)


# ── Step 6: Fix Data Types ──
df['Survived'] = df['Survived'].astype(bool)
df['Pclass']   = df['Pclass'].astype('category')
df['Sex']      = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')


# ── Step 7: Rename Columns ──
df.columns = df.columns.str.lower()
df.rename(columns={
    'sibsp': 'siblings_spouses',
    'parch': 'parents_children'
}, inplace=True)

# last  feature engineering  new useful column from existing data

df['family_size'] = df['siblings_spouses'] + df['parents_children'] + 1
df['is_alone'] = (df['family_size'] == 1).astype(int)
df['title'] = df['name'].str.extract(r' ([A-Za-z]+)\.')

# ── Step 9: Final Check ──
print(f"\nCleaned shape: {df.shape}")
print(df.isnull().sum())
print(df.dtypes)
print(df.head())
```
