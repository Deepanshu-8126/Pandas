```python
# real dataset work titanic
import pandas as pd
df = pd.read_csv('/content/Titanic-Dataset.csv')

# basic look
df.shape
df.head()
print(df.isnull().sum())

# Step 3: Basic stats
print(df.describe())

# Step 4: Specific column analysis
print(df['Survived'].value_counts())
print(df['Pclass'].value_counts())

# step 5- survival rates
survival_rate = df['Survived'].mean() * 100
print(f'Survival Rate: {survival_rate:.2f}%')  # 38.38%

#step 6- new column - age group
df['Age_Group'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')
print(df[['Name','Age','Age_Group']].head(10))
```
