```python
import pandas as pd
import numpy as np

# series  ( 5 methods to create series )
marks = pd.Series([85,90,78,92,88])
print(marks)
```

```python
# custom index
marks = pd.Series([85,90,78,92,88], index = ['Math','Science','English','Hindi', 'sst'])
print(marks)
```

```python
# dictionary
prices = {'Apple': 150, 'Mango': 80, 'Banana': 40}
s = pd.Series(prices )
s
```

```python
# from scalar
s = pd.Series(5,index=[0,1,2,3])
print(s)
```

```python
# numpy array
arr = np.array([10,20,30,40])
s = pd.Series(arr)
s
```

```python
# topic 1.3 indexing and access
marks = pd.Series([85, 90, 78, 92, 88],
                  index=['Math','Science','English','Hindi','SST'])

print(marks[0])
print(marks[2])

# custom label
print(marks['Math'])

# mulitple labels double brackets
marks[['Math','Hindi']]
```

```python
# series attributes
print(marks.values)
print(marks.dtype)
print(marks.size)
print(marks.name) # series name
print(marks.index)
```

```python
# Topic : series operations
# note : operation element wise
s = pd.Series([10, 20, 30, 40, 50])
print(s+5) # add 5 in all element
print(s*2)
print(s**2) # square of each
print(s / 10 )  # divide each by 10
print(s % 3)
```

```python
# two series operations
s1 = pd.Series([10, 20, 30], index=['a','b','c'])
s2 = pd.Series([5,  10, 15], index=['a','b','c'])
print(s1+s2)
```

```python
# mismatches - nan comes
s1 = pd.Series([10, 20], index=['a','b'])
s2 = pd.Series([5,  15], index=['a','c']) # c is diffrent
print(s1+s2)

# a matched

# nan - not a number - matching index not identify
```

```python
# Comparison
s = pd.Series([85, 90, 78, 92, 88])
print(s>85)
print(s[s>85]) # filter only that elements who are greater than 85
```

```python
# Topic : Series Important methods
marks.describe()
marks.sum()
marks.mean()
marks.min()
marks.max()
marks.idxmin()
marks.idxmax() # index of max
marks.sort_values()
marks.sort_index() # index to sort
marks.value_counts() # unique value counts
marks.nunique() #  unnique value  count
marks.unique() # values return
marks.isnull()
marks.isnull().sum() # null values sum
marks.dropna()  # null values remove
#marks.fillna()  # null values replace
#marks.apply(func) # function at every element
```

```python
import pandas as pd
import numpy as np
scores = pd.Series([85, 92, np.nan, 78, 95, 88, np.nan, 70],
                   index=['Ali','Raj','Sam','Tom','Priya','Dev','Mia','Zoe'])

scores.describe()
print(scores.isnull().sum()) # 2 null values

filled = scores.fillna(scores.mean())
filled

# top scorer check
print('Top scorer :', scores.idxmax(), scores.max())

scores[scores>85]


# apply() - grade assign
def get_grade(x):
  if pd.isna(x): return 'N/a',
  elif x >= 90:  return 'A',
  elif x >= 80:  return 'B',
  else:           return 'C'

print(scores.apply(get_grade))
```
