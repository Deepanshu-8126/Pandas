```python
student_dict={
  "iq":[100,90,120,80,0,0],
  "marks":[80,70,100,50,0,0],
  "package":[10,7,14,2,0,0]
}
students=pd.DataFrame(student_dict)
print(students)

# lets check in any column any duplicated item / value
print(students.duplicated().sum()) # 1 item is duplicated
```

```python
# rename ( marks - change - percent and package change to - lpa )
# using columns
# column parameter ( old value and new value )


# note that are not permanent change original student data is as it is

# inplace = true permanent change
students.rename(columns={'marks':'percent','package':'lpa'},inplace=True)
print(students)
```

```python
# Mathematical functions
# sum mean median axis argument -- automatically do sum at every column
students.sum() # total sum of all columns
print(students)
# rows also aply
students.sum(axis=1)

# this axis parameter is applicable for all mathematical functions
# like mean , median etc
```

```python
students.mean()
students.min() # default here column wise minimum
students.min(1)
```

```python
students.max()
students.median()
students.std()
students.var()

# same as in series but here both axis operation perform row wise sum etc
```

```python
# example of student dictionary
student_dict={
  'name':['nitin','ankit','rupesh','risabh','amit','ankita'],
  "iq":[100,90,120,80,0,0],
  "marks":[80,70,100,50,0,0],
  "package":[10,7,14,2,0,0]
}

# any column serve as an index
# example --- using set_index
students=pd.DataFrame(student_dict)
print(students)
students.set_index('name',inplace=True) # inplace true permanent change
```
