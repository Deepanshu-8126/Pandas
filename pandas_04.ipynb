```python
# setup
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Titanic-Dataset.csv')
print(df.shape)  # 891 rows and 12 columns
print(df.isnull().sum()) # total count of null values



# age - 177 missing , cabin = 687 missing too much missing
```

```python
# 4.1 missing values(nan)
# nan - not a number cell is empty - there is no value

missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent.round(2))



"""Age        19.87%   ← almost 20% missing
Cabin      77.10%   ← 77% missing — almost useless!
Embarked    0.22%   ← negligible"""
```

```python
# true false per cell there is any null
df.isnull().head()  #first five rows return


# there is any null in the dataframe
print(df.isnull().any().any()) # true is only one also


# in how mnay rows there is null values
print(df.isnull().any(axis=1).sum()) # 708
```

Missing values handle 4 types
dropna
fillna
interpolation
domain knowledge to fill


```python
# dropna : rows column drop
df_clean = df.dropna()
print(df_clean.shape)  # 183 12 many rows remove


# specific column mey null so drop
df_clean = df.dropna(subset = ['Age'])
df_clean.shape  # 714 12

# if 2 and more than  null in rows then drop
df_clean =  df.dropna(thresh=10)  #miniumn 10 non values need
df_clean.shape  # 891

# column drop if very null
df_clean = df.dropna(axis=1)
df_clean.shape # 891 9


# rules : if 40 percent more than null valeus in ccolumn then column drop beeter
```

```python
# fillna  : fill the values

# specific value to fill 0 to fill
df['Age'].fillna(0,inplace=True) # bad practice

# mean to fill common
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Age'].head(10)


# median to fill
df['Age'].fillna(df['Age'].median(),inplace=True)
# median - values sort middle one take


# mode to fill  - for categorial data
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)


# Forward fill - first valeu comes from the null value
df['Age'].fillna(method='ffill', inplace=True)


# backward fill
df['Age'].fillna(method='bfill', inplace=True)
```

```python
# 3 way- interpolation
# estimate middle value mathematically time series data use

df['Age'].interpolate(method='linear',inplace = True)

# domian knowledge fill
df['Cabin'].fillna('Unknown', inplace=True)
```

```python
# 4.2 duplicates values
print(df.duplicated().sum()) # duplicate row

```

```python
# which row is duplicated
print(df[df.duplicated()])

# specific columns
print(df.duplicated(subset=['Name', 'Age']).sum())

# all duplicates show
print(df[df.duplicated(keep=False)])
```

```python
# drop duplicates
df.drop_duplicates()

# last occurence
df.drop_duplicates(keep='last')

# specific columns duplicates
df.drop_duplicates(subset=['Name','Ticket'])


df.drop_duplicates(inplace=True)
```

```python
# data type ( dtypes) ans as type()
print(df.dtypes)
```

```python
# astype - change the data type
# column data type change

# survived column it is integer64 but we need boolean

df['Survived'] = df['Survived'].astype(bool)
print(df['Survived'].dtype) # bool return



# pclass convert to category - memory save

df['Pclass'] = df['Pclass'].astype('category')


#str to integer

df['PassengerId'] = df['PassengerId'].astype(str)



df2 = pd.DataFrame({'Price': ['500', '1500', '800', '2000']})
df2['Price'] = df2['Price'].astype(int)
print(df2['Price'].dtype)
```
