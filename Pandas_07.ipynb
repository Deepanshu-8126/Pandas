```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/netflix_titles.csv')
print(df.shape)
df.columns.tolist()
print(df.head(3))
```

```python
print(df.isnull().sum())
```

```python
# 7.1 concat - stack add dataframe
# down or up add
# axis = 0 row and axis = 1 column

# netflix dataset under concat
# Netflix data Movies aur TV Shows mein baanto

# split
movies = df[df['type'] == 'Movie'].copy()
tv_shows = df[df['type'] == 'TV Show'].copy()

print(movies.shape)
print(tv_shows.shape)


# combined
combined = pd.concat([movies, tv_shows])
print(f"Combined:{combined.shape}")

combined.head(5)
```

```python
# concat important features
# ignore_iondex = True fresh index
combined = pd.concat([movies , tv_shows], ignore_index = True)

# keys — source track
combined = pd.concat([movies, tv_shows],
                     keys=['Movies', 'TV Shows'])
print(combined.loc['Movies'].head(3))
print(combined.loc['TV Shows'].head(3))
```

```python
# axis =1 side  add
basic_info = df[['show_id', 'type', 'title', 'release_year']]
extra_info = df[['rating', 'duration', 'listed_in']]

combined_cols = pd.concat([basic_info , extra_info], axis = 1)
print(combined_cols)
```

```python
# duplicates check iindex
try:
    pd.concat([movies, tv_shows], verify_integrity=True)
except ValueError as e:
    print(f"Error: {e}")
```

```python
# 7.2 merge()

# table createe
shows = df[['show_id', 'type', 'title', 'release_year']].copy()
# Table 2: Show details
details = df[['show_id', 'rating', 'duration',
              'country', 'listed_in']].copy()

# table 3 :

directors = df[['show_id', 'director']].dropna().copy()

```

```python
# 7.3 4 types of join
# inner join
# left join
# 3 - right join
# 4- outer join

# type 1 - inner join

result = pd.merge(shows, details, on='show_id', how='inner')
print(result.head(3))
```

```python
# left join
result_left = pd.merge(shows , directors , on = 'show_id', how = 'left')
print(result_left.shape)
print(result.head(8))
print(result_left.isnull().sum())
```

```python
# Right join
result_right = pd.merge(shows, directors,
                        on='show_id',
                        how='right')
```

```python
# outer join
result_outer = pd.merge(shows, directors,
                        on='show_id',
                        how='outer')
print(result_outer.shape)
print(result_outer.isnull().sum())
```
