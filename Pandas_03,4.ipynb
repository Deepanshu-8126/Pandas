 loc, iloc, Boolean Filtering & query()

```python
# loc - find by label
# iloc - position number to find
# bollean filtering - find by conditons
```

```python
# setup
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Titanic-Dataset.csv')
print(df.shape)
df.head()
```

```python
# 3.1 label based indexing
# label = index name  - default 0 ,1,2,3 thesee are the labels
# df.loc[row_label, column_label]


# case 1 single row
df.loc[0]


# case 2 - mulitple rows
df.loc[0:2]  # 0 1 2 three include
```

```python
# case 3 specifi columns + specific column
#  row 0 name and age only colum n
df.loc[0,'Name']

print(df.loc[0, 'Age'])


# multiple columns of first row 0
print(df.loc[0,['Name','Age','Survived']])
```

```python
# case 4 - rows range + column range
# rows 0 to 4 , column name to age
print(df.loc[0:4,'Name':'Age'])
```

```python
# case 5 - custom index with loc
df2 = df.set_index('Name')

# find with name
print(df2.loc['Braund, Mr. Owen Harris'])
```

```python
# 3.2 iloc : position based indexing
# -- nummber based loaction 0 - first 1- second

# case 1 - single row
df.iloc[0] # first row
df.iloc[-1] # last row
df.iloc[5]    # 6th row
```

```python
# case 2- row range
df.iloc[0:3] # exclude 0 1 2
df.iloc[5:10]
```

```python
# case 3- row + column position
df.iloc[0,3] # row , column
df.iloc[0,0:4] # columnss 0 to 3
df.iloc[0:5,0:4] # rows 0 to 4 and columns 0 to 3
```

```python
# case 4- specific rows + specific columns  fromm list
df.iloc[[1,2,4],[0,2,4]] # list under list
```

```python
# cae 5 - useful tricks
# first 5 rows all columns
df.iloc[:5, :]

# alll rows and first 3 columns
df.iloc[:,:3]

# lasat 3 rows and last 2 columns
df.iloc[-3:,-2:]
```

```python
# same result diffrent way
df.loc[0:2,'Name':'Sex']
df.iloc[0:3,2:5]
```

Boolean filtering

```python
condition = df['Age']>30
df.shape # 891 12
# dataframe filter
old_passengers = df[condition]
old_passengers.shape # filter whose age is 30+ returned filtering


```

```python
# single condition
# 1st class pasengers
first_class = df[df['Pclass']==1]
print(first_class.shape) # 216,12


# female passengers
females = df[df['Sex']=='female']
print(females.shape)


# fare > 100
rich = df[df['Fare']>100]
rich.shape


# survived passengers
survived = df[df['Survived']==1]
survived.shape #  324 12
```

```python
# multiple conditions AND (&), OR (|), NOT (~)
# female and survived
female_survived = df[(df['Sex']=='female') & df['Survived']==1]
female_survived

# 1 st class and age>50
rich_old = df[(df['Pclass']==1) & (df['Age']>50)]
rich_old
```
