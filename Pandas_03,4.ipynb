 loc, iloc, Boolean Filtering & query()

```python
# loc - find by label
# iloc - position number to find
# bollean filtering - find by conditons
```

```python
# setup
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Titanic-Dataset.csv')
print(df.shape)
df.head()
```

```python
# 3.1 label based indexing
# label = index name  - default 0 ,1,2,3 thesee are the labels
# df.loc[row_label, column_label]


# case 1 single row
df.loc[0]


# case 2 - mulitple rows
df.loc[0:2]  # 0 1 2 three include
```

```python
# case 3 specifi columns + specific column
#  row 0 name and age only colum n
df.loc[0,'Name']

print(df.loc[0, 'Age'])


# multiple columns of first row 0
print(df.loc[0,['Name','Age','Survived']])
```

```python
# case 4 - rows range + column range
# rows 0 to 4 , column name to age
print(df.loc[0:4,'Name':'Age'])
```

```python
# case 5 - custom index with loc
df2 = df.set_index('Name')

# find with name
print(df2.loc['Braund, Mr. Owen Harris'])
```

```python
# 3.2 iloc : position based indexing
# -- nummber based loaction 0 - first 1- second

# case 1 - single row
df.iloc[0] # first row
df.iloc[-1] # last row
df.iloc[5]    # 6th row
```

```python
# case 2- row range
df.iloc[0:3] # exclude 0 1 2
df.iloc[5:10]
```

```python
# case 3- row + column position
df.iloc[0,3] # row , column
df.iloc[0,0:4] # columnss 0 to 3
df.iloc[0:5,0:4] # rows 0 to 4 and columns 0 to 3
```

```python
# case 4- specific rows + specific columns  fromm list
df.iloc[[1,2,4],[0,2,4]] # list under list
```

```python
# cae 5 - useful tricks
# first 5 rows all columns
df.iloc[:5, :]

# alll rows and first 3 columns
df.iloc[:,:3]

# lasat 3 rows and last 2 columns
df.iloc[-3:,-2:]
```

```python
# same result diffrent way
df.loc[0:2,'Name':'Sex']
df.iloc[0:3,2:5]
```

Boolean filtering

```python
condition = df['Age']>30
df.shape # 891 12
# dataframe filter
old_passengers = df[condition]
old_passengers.shape # filter whose age is 30+ returned filtering


```

```python
# single condition
# 1st class pasengers
first_class = df[df['Pclass']==1]
print(first_class.shape) # 216,12


# female passengers
females = df[df['Sex']=='female']
print(females.shape)


# fare > 100
rich = df[df['Fare']>100]
rich.shape


# survived passengers
survived = df[df['Survived']==1]
survived.shape #  324 12
```

```python
# multiple conditions AND (&), OR (|), NOT (~)
# female and survived
female_survived = df[(df['Sex']=='female') & df['Survived']==1]
female_survived

# 1 st class and age>50
rich_old = df[(df['Pclass']==1) & (df['Age']>50)]
rich_old

# or - one condition satisfy
result = df[(df['Pclass']==1) | (df['Survived']==1)]
result

# not reverse the condition
males = df[~(df['Sex']=='female')]
# males passengers not female
```

```python
# .isin() - multiple values check - is in the list

# pclass 1 or 2 passengers
upper_class = df[df['Pclass'].isin([1,2])]
print(upper_class)

result = df[df['Embarked'].isin(['S', 'C'])]
print(result.shape)
```

```python
#.between range check
# between(a,b)

# age 20 and 30 between
young_adults = df[df['Age'].between(20,30)]
young_adults.shape


# fare 10 to 50
mid_fare = df[df['Fare'].between(10,50)]
mid_fare.shape
```

```python
# .str methods = text filtering
# str = string operations columnn apply

# in name .Mr one passengers
mr_passengers = df[df['Name'].str.contains('Mr')]
mr_passengers

# name in uppercase
df['Name_Upper']=df['Name'].str.upper()

#.str_lower

# length - str.len()

# first three characters
df['frst3']=df['Name'].str[:3]
```

```python
# filtering  + specific columns
# question - female passengers then only name agae fare

result = df[df['Survived']==1][['Name','Age','Fare','Sex']]
result

```

```python
# real analysis in titanic
female_survival = df[df['Sex'] == 'female']['Survived'].mean() * 100
print(f"{female_survival:.1f}%")

print("Male survival rate:")
male_survival = df[df['Sex'] == 'male']['Survived'].mean() * 100
print(f"{male_survival:.1f}%")


# in 1st class how many survived
print("1 st class survived")
first_class = df[(df['Pclass']==1) & (df['Survived']==1)]
print(f"1st class survivors {len(first_class)}")



# fare
max_fare_person = df.iloc[df['Fare'].idxmax()]
print(max_fare_person)

max_fare_person = df.loc[df['Fare'].idxmax(), ['Name', 'Fare', 'Pclass']]
print(max_fare_person)
```

```python
# 3.4 query
# normal filtering
df[df['Age']>30]


# query - very clean
df.query('Age>30')

#Multiple conditions
df.query('Age > 30 and Pclass == 1').head(5)

df.query('Sex == "female" and Survived == 1')


"""df.query('Age between 20 and 30')
df.query('Pclass in [1, 2]')"""

# variable use in query
min_age = 25
df.query('Age > @min_age').head(6)
```
